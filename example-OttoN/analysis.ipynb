## Step 1: Acquire DEM list

```python
if 1 in do_steps:
    print('Doing Step 1...')
    achydro.acquire_dem_list(arcticdem_index, refgeo_gpkg, dem_input)
else:
    print('Step 1 skipped')
```

## Step 2: Download Subset DEMs

```python
if 2 in do_steps:
    print('Doing Step 2...')
    achydro.download_subset_dems(dem_input, dem_subsets_folder, bitmask_subsets_folder, refgeo_gpkg, buffer_size)
else:
    print('Step 2 skipped')
```

## Step 3: Acquire ICESat-2 ATL-6 points

Before running this step, you'll need to specify the EarthData login credential using environment variables:

- EARTHDATA_USERNAME: username
- EARTHDATA_PASSWORD: password

```python
if 3 in do_steps:
    print('Doing Step 3...')
    achydro.download_is2_atl6_data(refgeo_gpkg, altimetry_subsets_folder)
else:
    print('Step 3 skipped')
```

## Step 4: Select ICESat-2 ATL-6 points over good bedrock

```python
if 4 in do_steps:
    print('Doing Step 4...')
    achydro.filter_is2_atl6_data(altimetry_subsets_folder, altimetry_processed_folder, rgi_index, refgeo_gpkg, 
                                 working_epsg, off_ice_buffer, selected_elev_file, 
                                 if_additional_exclusion=if_additional_exclusion,
                                 additional_exclusion_file=additional_exclusion_gpkg)
else:
    print('Step 4 skipped')
```

## Step 5: Align every DEMs with ICESat-2 reference points, and calcuate uncertainty

```python
if 5 in do_steps:
    print('Doing Step 5...')
    achydro.align_dems(refgeo_gpkg, buffer_size, selected_elev_file, 
                       dem_subsets_folder, dem_processed_folder, bitmask_subsets_folder)
else:
    print('Step 5 skipped')
```

## Step 6: Analyze the processed DEMs and make manifest

```python
if 6 in do_steps:
    print('Doing Step 6...')
    achydro.make_dem_manifest(dem_summary, dem_processed_folder, dem_input_processed)
else:
    print('Step 6 skipped')
```

## Step 7: Stack all DEMs

```python
if 7 in do_steps:
    print('Doing Step 7...')
    pile = achydro.stack_dems(tag, refgeo_gtif, carst_inipath, dem_input_processed, refdate, evmd_threshold)
else:
    print('Step 7 skipped')
```

## Step 8: Visualization of the stacked elevations

```python
if 8 in do_steps:
    print('Doing Step 8...')
    if 'pile' not in locals():
        pile = DemPile()
        pile.read_config(carst_inipath)
        pile.load_pickle()
    pile.viz(evmd_threshold=evmd_threshold, min_samples=min_samples, reg_method='sigmoid', 
             use_bitmask_only=False, use_evmd_only=False)
else:
    print('Step 8 skipped')
```

## Step 9: Fit a dh/dt model, determine maximum elevation change and timing

```python
if 9 in do_steps:
    print('Doing Step 9...')
    if 'pile' not in locals():
        pile = DemPile()
        pile.read_config(carst_inipath)
        pile.load_pickle()
    drain_dh_map, drain_t_map = achydro.dhdt_fit_sigmoid(pile, tag, final_output_tag, 
                                                         nproc=nproc, evmd_threshold=evmd_threshold, k_bounds=k_bounds)
else:
    print('Step 9 skipped')
```

## Step 10: Result visualization

```python
import numpy as np
from carst.libdhdt import sigmoid_reg as smr
from datetime import timedelta
import rasterio
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from cmcrameri import cm

if 10 in do_steps:
    print('Doing Step 10...')
    if 'pile' not in locals():
        pile = DemPile()
        pile.read_config(carst_inipath)
        pile.load_pickle()

    # You have to at least run step 7 once to get the following file
    quick_topography_path = f'{pile.dhdtprefix}_median_topo.tif'
    toporaster = rasterio.open(quick_topography_path)
    topo = toporaster.read(1)
    left, bottom, right, top = toporaster.bounds
    # img_extent = (left / 1000, right / 1000, bottom / 1000, top / 1000)
    img_extent = (left, right, bottom, top)

    # You have to at least run step 9 once to the following file
    water_dh_path = f'{pile.dhdtprefix}_{final_output_tag}_dh.tif'
    water_dhraster = rasterio.open(water_dh_path)
    water_dh = water_dhraster.read(1)

    if 'gdf_rgi' not in locals():
        gdf_rgi = achydro.read_zipped_shapefile(rgi_index, working_epsg)
    if 'ref_h' not in locals():
        ref_h = np.loadtxt(selected_elev_file, skiprows=1, delimiter=",")

    exitstate, evmd_lbl = pile.ts[row, col].do_evmd(eps=evmd_threshold, min_samples=min_samples)
    bitmask_lbl = pile.ts[row, col].bitmask_labels
    good_idx = np.logical_and(bitmask_lbl == 0, evmd_lbl >= 0)
    arr = pile.ts[row, col]
    x = arr.get_date()
    y = arr.get_value()
    ye = arr.get_uncertainty()
    x_date = np.array([pile.refdate + timedelta(days=i) for i in x])

    sampleloc_x = (col / pile.ts.shape[1]) * (right - left) + left
    sampleloc_y = top - (row / pile.ts.shape[0]) * (top - bottom)
    
    fig, axs = plt.subplots(2, 2, figsize=(10, 10), sharey=True)
    gs = axs[0, 1].get_gridspec()
    for ax in axs[1, :]:
        ax.remove()
    axbottom = fig.add_subplot(gs[1, :])
    
    h_im = axs[0, 0].imshow(topo, extent=img_extent, cmap='gist_earth')
    axs[0, 0].plot(ref_h[:, 0], ref_h[:, 1], 'o', color='xkcd:salmon', markersize=2.5, markeredgecolor='xkcd:black', markeredgewidth=0.3)
    dh_im = axs[0, 1].imshow(water_dh, extent=img_extent, cmap=cm.lapaz, clim=(-60, 0))

    for ax in axs[0, :]:
        gdf_rgi.plot(ax=ax, facecolor='none', edgecolor='xkcd:black', linewidth=1)
        ax.plot(sampleloc_x, sampleloc_y, '*', markersize=7, markeredgewidth=0.3, markeredgecolor='xkcd:black', color='xkcd:violet')

    axs[0, 0].set_xlim(left, right)
    axs[0, 0].set_ylim(bottom, top)
    axs[0, 1].set_xlim(left, right)
    
    fig.colorbar(h_im)
    fig.colorbar(dh_im)

    axbottom.errorbar(x_date, y, yerr=ye * 2, linewidth=2, fmt='k.')
    # np.set_printoptions(precision=3)
    # np.set_printoptions(suppress=True)

    x_good = x[good_idx]
    y_good = y[good_idx]
    ye_good = ye[good_idx]
                
    x_pred, y_pred, sigmoid_height, sigmoid_height_stderr, sigmoid_timing, exitstate = smr(x_good, y_good, ye=ye_good, k_bounds=k_bounds)
    axbottom.set_title(f'Sigmoid height (2-sigma) = {sigmoid_height:.2f}{2 * sigmoid_height_stderr:.2f}' )
    x_pred_date = [pile.refdate + timedelta(days=i) for i in x_pred]
    axbottom.plot(x_pred_date, y_pred, color='g', linewidth=2, zorder=20)
    x_good_date = [pile.refdate + timedelta(days=i) for i in x_good]
    axbottom.plot(x_good_date, y_good, 's', color='k', markersize=7, zorder=5)
    if not np.isnan(sigmoid_timing):
        axbottom.axvline(refdate + timedelta(days=sigmoid_timing), linestyle='--', color='xkcd:teal')


    bitmask_colorcodes = ['xkcd:lilac', 'xkcd:gray', 'xkcd:light blue',  'xkcd:blue', 
                          'xkcd:light yellow',  'xkcd:yellow', 'xkcd:gold', 'xkcd:brown']
    bitmask_comments = ['Good', 'Edge', 'Water', 'Water+Edge',              
                        'Cloud', 'Cloud+Edge', 'Cloud+Water', 'Cloud+Edge+Water']
    for selected_bit, colorcode in zip([0, 1, 2, 3, 4, 5, 6, 7], bitmask_colorcodes):
        selected_group_index = bitmask_lbl == selected_bit
        axbottom.plot(x_date[selected_group_index], y[selected_group_index], '.', color=colorcode, 
                    markersize=12, markeredgewidth=1, markeredgecolor='k', zorder=selected_bit + 10)


    axbottom.xaxis.set_major_locator(mdates.YearLocator())
    axbottom.xaxis.set_minor_locator(mdates.MonthLocator())
    plt.setp(axbottom.get_xticklabels(), rotation=-30, ha='left')
    axbottom.grid(axis='x', linestyle='--', color='xkcd:light grey')
    axbottom.set_xlabel('data[{}, {}]'.format(row, col))
    axbottom.set_ylabel('height (m)')

    fig.patch.set_facecolor('xkcd:white')
    fig.savefig(f'{pile.dhdtprefix}_{final_output_tag}_viz.png')
else:
    print('Step 10 skipped')
```
